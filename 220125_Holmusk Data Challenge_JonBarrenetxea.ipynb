```python
#######################################################################################################################
###############################################CLEANING################################################################
#######################################################################################################################

###Import event duration and patient characteristics datasets

f1 = pd.read_csv("Event_duration.csv")
f2 = pd.read_csv("Patient_characteristics.csv")

display(f1)
print(f1.info())

display(f2)
print(f2.info())
```

```python
###Merge event duration and patient characteristics datasets

data_survival=pd.merge(f1, f2)

display(data_survival)
```

```python
###Rename variables

data_survival = data_survival.rename(columns = {'Bleeding_event (1=event, 0=censored)': 'bleeding_event'}, inplace = False)
```

```python
###Hot encoding to turn oject data into numeric

## Get one hot encoding of columns 

object_cols=['treatment_variable','other_drugs_1','other_drugs_2','other_drugs_3','other_drugs_4','other_drugs_5','other_drugs_6','other_drugs_7','other_drugs_8',
             'diagnosis_1','diagnosis_2','diagnosis_3','diagnosis_4','diagnosis_5','diagnosis_6','diagnosis_7','diagnosis_8','diagnosis_9','diagnosis_10','diagnosis_11','diagnosis_12','diagnosis_13','diagnosis_14','diagnosis_15']


data_survival_hot = pd.get_dummies(data_survival, columns = object_cols)
display(data_survival_hot)
print(data_survival_hot.info())

##Keep only those columns ending in '_yes' for a more parsimonious dataset, and to avoid colinearity issues in downstream regression models

data_survival_final = data_survival_hot[data_survival_hot.columns.drop(list(data_survival_hot.filter(regex='_No')))]
print(data_survival_final.info())
```

```python
###Create number of comorbidities variable (sum of all other diagnosis columns)
data_survival_final["number_comorbidities"]=data_survival_final["diagnosis_1_Yes"]+data_survival_final["diagnosis_2_Yes"]+data_survival_final["diagnosis_3_Yes"]+data_survival_final["diagnosis_4_Yes"]+data_survival_final["diagnosis_5_Yes"]+data_survival_final["diagnosis_6_Yes"]+data_survival_final["diagnosis_7_Yes"]+data_survival_final["diagnosis_8_Yes"]+data_survival_final["diagnosis_9_Yes"]+data_survival_final["diagnosis_10_Yes"]+data_survival_final["diagnosis_11_Yes"]+data_survival_final["diagnosis_12_Yes"]+data_survival_final["diagnosis_13_Yes"]+data_survival_final["diagnosis_14_Yes"]+data_survival_final["diagnosis_15_Yes"]
print(data_survival_final["number_comorbidities"].value_counts())

##Collapse those with more than 5 conditions in one category
data_survival_final.loc[data_survival_final["number_comorbidities"]>5,"number_comorbidities"] = 6
print(data_survival_final["number_comorbidities"].value_counts())



```

```python
###Create number of other drugs (sum of all other diagnosis columns)
data_survival_final["number_drugs"]=data_survival_final["other_drugs_1_Yes"]+data_survival_final["other_drugs_2_Yes"]+data_survival_final["other_drugs_3_Yes"]+data_survival_final["other_drugs_4_Yes"]+data_survival_final["other_drugs_5_Yes"]+data_survival_final["other_drugs_6_Yes"]+data_survival_final["other_drugs_7_Yes"]+data_survival_final["other_drugs_8_Yes"]
print(data_survival_final["number_drugs"].value_counts())

##Collapse those with more than 3 drugs in one category
data_survival_final.loc[data_survival_final["number_drugs"]>3,"number_drugs"] = 4
print(data_survival_final["number_drugs"].value_counts())
```

```python
###Data type inspection
print(data_survival_final.info())
```

```python
###Drop irrelevant variables for analysis 

##Drop drug_A binary variable to avoid colinearity issues in downstream regressions
data_survival_final.drop('treatment_variable_Drug_A', inplace=True, axis=1)

##Drop patient_id from analysis dataset
data_survival_final.drop('patient_id', inplace=True, axis=1)
```

```python
###Check proportionality of hazards assumption

cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event')

cph.print_summary()

cph.check_assumptions(data_survival_final, p_value_threshold=0.05, show_plots=True) #Verify proportionality of hazards

cph.plot()


```

```python
#lab_5 and Diag_Score 2 do not fit proportionality of hazards assumption
```

```python
###Base proportional hazards model 

cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event', formula='treatment_variable_Drug_B + age + sex')
cph.print_summary() 
cph.plot()
```

```python
#Results: According to the base model (adjusted for age and sex), those taking Drug B are 10% more likely to have a bleeding event than those taking drug A
```

```python
###Estimate propensity scores using logistic regression (requires categorical data to be in dummy mode)
from sklearn.linear_model import LogisticRegression

T = 'treatment_variable_Drug_B'
Y = 'bleeding_event'
X = data_survival_final.columns.drop([T, Y]) #Confounders, every other variable except the treatment and the event

ps_model = LogisticRegression(C=1e6, max_iter=10000).fit(data_survival_final[X], data_survival_final[T])

data_ps = data_survival_final.assign(propensity_score=ps_model.predict_proba(data_survival_final[X])[:, 1])

data_ps[["treatment_variable_Drug_B", "bleeding_event", "propensity_score"]].head()
```

```python
#Check overlap of propensity scores for those taking Drug A  and Drug B
sns.distplot(data_ps.query("treatment_variable_Drug_B==0")["propensity_score"], kde=False, label="Drug A")
sns.distplot(data_ps.query("treatment_variable_Drug_B==1")["propensity_score"], kde=False, label="Drug B")
plt.title("Distribution of Propensity Scores")
plt.legend();
```

```python
#There is some overlap between the population treated with Drug A and Drug B, but the distribution is somewhat unbalanced.
#This may cause some bias because of a) the high weight of the participants with extreme values of 1.0 or 0.0 and b) more difficult to find matching pairs for treatment effect comparison

```

```python
#####Run PH model accounting for propensity scores

cph = CoxPHFitter()
cph.fit(data_ps, duration_col='duration_in_years', event_col='bleeding_event', formula= "treatment_variable_Drug_B + propensity_score")
cph.print_summary() 
cph.plot()

```

```python
#After balancing the dataset, we find no evidence of differences between drugs A and B for the occurence of bleeding events (HR=1.02, 95%CI:0.94-1.10)
```

```python
### 'Simulate an RCT' by matching participants with the same propensity score matching
cm = CausalModel(
    Y=data_ps["bleeding_event"].values, 
    D=data_ps["treatment_variable_Drug_B"].values, 
    X=data_ps[["propensity_score"]].values
)

cm.est_via_matching(matches=1, bias_adj=True)

print(cm.estimates)
```

```python
#Interpretation: there is no difference between Drug A and Drug B as to the occurrence of bleeding events
```

```python
##Model 1: Adjusted for age + gender + Diagnosis Scores
cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event', formula= "treatment_variable_Drug_B + age + sex + Diag_Score_1 + Diag_Score_2")
cph.print_summary() 
cph.plot()
```

```python
# Model 1: Those treated with Drug B are at 11% higher risk of developing bleeding events compared to those treated with Drug A
```

```python
# Model2: Adjusted for age + gender+ diagnosis scores + other drugs
cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event', formula= "treatment_variable_Drug_B + age + sex + Diag_Score_1 + Diag_Score_2 +other_drugs_1_Yes + other_drugs_2_Yes + other_drugs_3_Yes + other_drugs_4_Yes + other_drugs_5_Yes + other_drugs_6_Yes + other_drugs_7_Yes + other_drugs_8_Yes")
cph.print_summary() 
cph.plot()
```

```python
#After adjusting for other_drugs, the treatment effect for Drug B becomes non significant. 'Other_drugs_2' is the strongest confounder.
```

```python
# Model3: Adjusted for age + gender + diagnosis scores+ Comorbidities
cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event', formula= "treatment_variable_Drug_B + age + sex + Diag_Score_1 + Diag_Score_2 + diagnosis_1_Yes + diagnosis_2_Yes + diagnosis_3_Yes + diagnosis_4_Yes + diagnosis_5_Yes + diagnosis_6_Yes + diagnosis_7_Yes + diagnosis_8_Yes + diagnosis_9_Yes + diagnosis_10_Yes + diagnosis_11_Yes + diagnosis_12_Yes + diagnosis_13_Yes + diagnosis_14_Yes +diagnosis_15_Yes ")
cph.print_summary() 
cph.plot()
```

```python
# Similarly, when adjusting for different comorbidities, the treatment effect of Drug B fades away. 'Diagnosis_12' is by far the strongest confounder
```

```python
#Model4: Adjusted for age + gender + diagnosis scores+ lab results
cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event', formula= "treatment_variable_Drug_B + age + sex + Diag_Score_1 + Diag_Score_2 + lab_1 + lab_2 + lab_3 + lab_4 + lab_5 + lab_6 + lab_7 ")
cph.print_summary() 
cph.plot()
```

```python
# After accounting for lab test results, the treatment effect of Drug B fades away. Lab_1 and lab_7 are the strongest confounders
```

```python
#Fully adjusted model with all covariates
cph = CoxPHFitter()
cph.fit(data_survival_final, duration_col='duration_in_years', event_col='bleeding_event')
cph.print_summary() 
cph.plot()
```

```python
# After full multivariable adjustment, we conclude that there is no difference between Drugs A and B for the occurrence of bleeding events. 
# Any difference between Drugs A and B is likely to be confounded by the uptake of other drugs, other comorbidities and lab test results
# One particular confounders stands out for its effect size, namely: diagnosis_12
```
